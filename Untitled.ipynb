# Import Libs

```python
import numpy as np
import pandas as pd
```

# Pre-processing

```python
job_postings_data = pd.read_csv("postings.csv")
job_postings_data
```

## Kiểm tra duplicate hay null ở job_id

```python
duplicates = job_postings_data['job_id'].duplicated()
num_duplicates = duplicates.sum()
num_duplicates
```

```python
job_postings_data['job_id'].isnull().sum()
```

# Merge vào bảng benifits và loại bỏ cột inffered vì không sử dụng

```python
job_benefits_data = pd.read_csv("jobs/benefits.csv")
job_benefits_data = job_benefits_data.drop('inferred', axis=1)
job_benefits_data = job_benefits_data.groupby('job_id')['type'].agg(lambda x: ', '.join(x)).reset_index()
```

```python
job_postings_data = job_postings_data.merge(job_benefits_data, on="job_id", how="left")
job_postings_data
```

# Tương tự merge bảng job_skills vào

```python
job_skills_data = pd.read_csv("jobs/job_skills.csv")
job_skills_data = job_skills_data.groupby('job_id')['skill_abr'].agg(lambda x: ', '.join(x)).reset_index() 

job_postings_data = job_postings_data.merge(job_skills_data, on="job_id", how="left")
job_postings_data
```

# Đã xử lý xong về Job, bây giờ chuyển sang xử lý Company

```python
company_data = pd.read_csv("companies/companies.csv")

company_industries_data = pd.read_csv("companies/company_industries.csv")
company_industries_data = company_industries_data.groupby('company_id')['industry'].agg(lambda x: ', '.join(x)).reset_index()

company_specialities_data = pd.read_csv("companies/company_specialities.csv")
company_specialities_data = company_specialities_data.groupby('company_id')['speciality'].agg(lambda x: ', '.join(x)).reset_index()

employee_counts_data = pd.read_csv("companies/employee_counts.csv")
employee_counts_data = employee_counts_data.groupby('company_id')['time_recorded'].max().reset_index()

company_data = company_data.merge(company_industries_data, on="company_id", how="left")
company_data = company_data.merge(company_specialities_data, on="company_id", how="left")
company_data = company_data.merge(employee_counts_data, on="company_id", how="left")
company_data
```

# Tiếp tục merge với Job Postings và Company. Ta sẽ có data về công ty đăng bài trên linkedln

```python
merged_data = job_postings_data.merge(company_data, on="company_id", how="left")
merged_data
```

```python
merged_data.isnull().sum()
```

# Loại bỏ các cột, để lại các cột để phân tích và đổi tên cột cho dễ đọc hiểu

```python
choosen_column = [
    'job_id',
    'company_name',
    'title',
    'max_salary',
    'pay_period',
    'location',
    'company_id',
    'views',
    'med_salary',
    'min_salary',
    'formatted_work_type',
    'applies',
    'original_listed_time',
    'remote_allowed',
    'application_type',
    'expiry',
    'skill_abr',
    'company_size',
    'state',
    'city',
    'industry',
    'speciality',
    'time_recorded'
]
merged_data = merged_data[choosen_column]
merged_data
```

# Bỏ các hàng nơi mà có company_name và speciality là NaN hoặc null

```python
pretty_column_name = {'title': 'job_title','formatted_work_type': 'work_type','original_listed_time': 'listed_time'}
data = merged_data.rename(columns=pretty_column_name)
data.dropna(subset=['company_name', 'speciality'], inplace=True)
data
```

# Điền vào các chỗ có giá trị NaN ở cột med_salary, min_salary, max_salary với giá trị điền vào là giá trị trung bình của cột med_salary, min_salary, max_salary

```python
data['med_salary'] = data['med_salary'].fillna(data['med_salary'].mean())
data['min_salary'] = data['min_salary'].fillna(data['min_salary'].mean())
data['max_salary'] = data['max_salary'].fillna(data['max_salary'].mean())
data
```

# Chuyển đổi UNIX TIME sang kiểu mm/dd/yyyy

```python
data['time_recorded'] = pd.to_datetime(data['time_recorded'], unit='s', errors='coerce').dt.strftime('%m/%d/%Y')

# Kiểm tra kết quả
print(data['time_recorded'].head(10))
```

# Điền vào chỗ còn trống của các cột pay_period, applies, remote_allowed, company_size, views với giá trị lần lượt là 'MONTHLY', 0, 0, 0, 0

```python
data['pay_period'] = data['pay_period'].fillna('MONTHLY')
data['applies'] = data['applies'].fillna(0)
data['remote_allowed'] = data['remote_allowed'].fillna(0)
data['company_size'] = data['company_size'].fillna(0)
data['views'] = data['views'].fillna(0)
data
```

```python
try:
    data.to_csv("cleaned-data.csv", index=False)
except Exception as e:
    print(f"An error occurred: {e}")
```
